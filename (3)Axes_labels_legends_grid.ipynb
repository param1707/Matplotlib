# **Axes,Labels,legends and grid in Matplotlib**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend()
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="upper right")
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best")
```

```python
days=[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True)
plt.grid()
```
